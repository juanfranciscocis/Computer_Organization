### Fibonacci

```
section .data
    fibonacci db 0, 1   ; Initialize the first two Fibonacci numbers

section .text
    global _start

_start:
    mov ecx, 2   ; Counter for Fibonacci number generation
    mov ebx, 1   ; Previous Fibonacci number
    mov eax, 1   ; Current Fibonacci number

fibonacci_loop:
    add eax, ebx   ; Calculate the next Fibonacci number
    mov ebx, eax   ; Update the previous Fibonacci number
    inc ecx        ; Increment the counter

    cmp ecx, 50    ; Check if we have reached the 50th Fibonacci number
    jne fibonacci_loop

    ; The 50th Fibonacci number is stored in EAX register
    ; You can print it or store it in memory for further use

    ; Exit the program
    mov eax, 1     ; System call number for exit
    xor ebx, ebx   ; Exit code 0
    int 0x80       ; Perform the system call


```

### Year leap?

```
section .data
    year_prompt db "Enter the year: ", 0
    leap_message db "Leap year!", 0
    not_leap_message db "Not a leap year.", 0

section .bss
    year_input resw 1

section .text
    global _start

_start:
    ; Prompt the user to enter the year
    mov eax, 4
    mov ebx, 1
    mov ecx, year_prompt
    mov edx, 16
    int 0x80

    ; Read the year from user input
    mov eax, 3
    mov ebx, 0
    mov ecx, year_input
    mov edx, 2
    int 0x80

    ; Convert the input to a number
    movzx eax, word [year_input]

    ; Check if it is a leap year
    mov edx, 0
    mov ebx, 4
    div ebx
    cmp edx, 0
    jne not_leap

    mov edx, 0
    mov ebx, 100
    div ebx
    cmp edx, 0
    je not_leap

    mov edx, 0
    mov ebx, 400
    div ebx
    cmp edx, 0
    je leap

not_leap:
    ; Print "Not a leap year."
    mov eax, 4
    mov ebx, 1
    mov ecx, not_leap_message
    mov edx, 17
    int 0x80
    jmp exit

leap:
    ; Print "Leap year!"
    mov eax, 4
    mov ebx, 1
    mov ecx, leap_message
    mov edx, 10
    int 0x80

exit:
    ; Exit the program
    mov eax, 1
    xor ebx, ebx
    int 0x80

```

### Calculate Average

```
section .data
    array dw 10, 20, 30, 40, 50   ; Example array of integers
    array_size equ ($ - array) / 2   ; Calculate the size of the array
    sum dw 0   ; Variable to store the sum of the array elements
    average dw 0   ; Variable to store the average

section .text
    global _start

_start:
    ; Calculate the sum of the array elements
    mov cx, array_size   ; Counter for array traversal
    mov si, 0   ; Initialize the array index

sum_loop:
    add ax, [array + si]   ; Add the array element to the sum
    add si, 2   ; Increment the array index by 2 bytes (word size)
    loop sum_loop

    mov [sum], ax   ; Store the sum in the 'sum' variable

    ; Calculate the average
    xor edx, edx   ; Clear the edx register
    mov ax, [sum]   ; Load the sum into the ax register
    mov bx, array_size   ; Load the array size into the bx register
    div bx   ; Divide the sum by the array size

    mov [average], ax   ; Store the average in the 'average' variable

    ; Print the average
    mov eax, 4   ; System call number for writing to standard output
    mov ebx, 1   ; File descriptor for standard output
    mov ecx, average   ; Pointer to the average variable
    mov edx, 2   ; Number of bytes to write (assuming average is a word)
    int 0x80   ; Perform the system call

    ; Exit the program
    mov eax, 1   ; System call number for exit
    xor ebx, ebx   ; Exit code 0
    int 0x80   ; Perform the system call

```

### Celsius <-> Fahrenheit

```
section .data
    celsius_prompt db "Enter temperature in Celsius: ", 0
    fahrenheit_prompt db "Enter temperature in Fahrenheit: ", 0
    converted_result db "Converted temperature: ", 0

section .bss
    celsius_input resw 1
    fahrenheit_input resw 1
    result resw 1

section .text
    global _start

_start:
    ; Prompt the user to enter temperature in Celsius
    mov eax, 4
    mov ebx, 1
    mov ecx, celsius_prompt
    mov edx, 29
    int 0x80

    ; Read the temperature in Celsius from user input
    mov eax, 3
    mov ebx, 0
    mov ecx, celsius_input
    mov edx, 2
    int 0x80

    ; Convert Celsius to Fahrenheit
    movzx eax, word [celsius_input]
    imul eax, 9
    idiv 5
    add eax, 32
    mov [result], ax

    ; Print the converted temperature
    mov eax, 4
    mov ebx, 1
    mov ecx, converted_result
    mov edx, 21
    int 0x80

    mov eax, 4
    mov ebx, 1
    mov ecx, result
    mov edx, 2
    int 0x80

    ; Prompt the user to enter temperature in Fahrenheit
    mov eax, 4
    mov ebx, 1
    mov ecx, fahrenheit_prompt
    mov edx, 33
    int 0x80

    ; Read the temperature in Fahrenheit from user input
    mov eax, 3
    mov ebx, 0
    mov ecx, fahrenheit_input
    mov edx, 3
    int 0x80

    ; Convert Fahrenheit to Celsius
    movzx eax, word [fahrenheit_input]
    sub eax, 32
    imul eax, 5
    idiv 9
    mov [result], ax

    ; Print the converted temperature
    mov eax, 4
    mov ebx, 1
    mov ecx, converted_result
    mov edx, 21
    int 0x80

    mov eax, 4
    mov ebx, 1
    mov ecx, result
    mov edx, 2
    int 0x80

    ; Exit the program
    mov eax, 1
    xor ebx, ebx
    int 0x80

```

### 3. Implement a function named `abs_diff` that calculates the absolute value of the difference of two inputs `a` and `b` (i.e., |a-b|), and get the assembly code output

```
section .text
    global abs_diff

abs_diff:
    push ebp
    mov ebp, esp
    sub esp, 4   ; Reserve space for local variables

    mov eax, [ebp+8]   ; Move 'a' into EAX register
    sub eax, [ebp+12]  ; Subtract 'b' from EAX
    cdq                ; Sign-extend EAX to EDX:EAX
    xor edx, eax       ; Get the absolute value in EDX:EAX

    mov [ebp-4], edx   ; Store the result in the stack

    mov eax, [ebp-4]   ; Move the result to EAX
    leave
    ret
```
